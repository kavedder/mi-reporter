```python
from bs4 import BeautifulSoup
import requests
import re
from html import unescape

MI_REPORTER_BASE = 'https://www.mi-reporter.com'
SECTIONS = ['news', 'sports', 'business']
HEADERS = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    
def get_article_text(url, author='Andy Nystrom'):
    html = requests.get(url, headers=HEADERS).text
    soup = BeautifulSoup(html)
    body = soup.body

    this_author = body.find('div', {'class': 'latest-by'}).a.text
    if this_author != author:
        return

    title = body.find('h2', {'class': 'entry-title'}).text
    text = body.find('div', {'class': 'entry-content'}).text
    return title, text

def find_in_text(text, regex):
    text_lower = text.lower()
    return re.findall(regex, text)


def clean_author(author, max_len=50):
    if not author:
        return author
    
    text = author
    
    # these are just weird
    # Matt Phelps’ email=’Matt.Phelps@mi-reporter.com
    # By Mercer Island resident Parisa Ardemani works with a cheetah in South Africa...
    # By Linda Williams Rorem While many young male ...
    # By John Stenberg Special to the Mercer Island Reporter...
    # By Joel Willits Bellevue Reporter...
    # By Proposal to re-name Tract A to Wildwood Park heads to City Council in September’ email=’...
    # By BY CARRIE RODRIGUEZ Mercer Island Reporter Editor
    text = text.replace('’ email=’', ' ')
    found = re.search(r'By Mercer Island resident (\w+ \w+) ', text)
    if found:
        return found[1]
        
    if 'Linda Williams Rorem' in text:
        text = 'Linda Williams Rorem'
    elif 'CARRIE RODRIGUEZ' in text:
         text = 'Carrie Rodriguez'
    elif 'Proposal to re-name Tract A to Wildwood Park heads to City Council in September' in text:
        text = 'Proposal to re-name Tract A to Wildwood Park heads to City Council in September'
    elif len(text) > max_len:
        found = re.search(r'By (\w+ \w+) ', text)
        if found:
            text = found[1]
    
    # unescape html
    text = unescape(text)
    
    # replace spaces
    text = re.sub(r'[\s\n\r]+', ' ', text, flags=re.MULTILINE).strip()
    
    # remove "by"-line lol
    text = re.sub(r'^by ', '', text, flags=re.IGNORECASE).strip()
    
    # remove email addresses
    EMAIL_RE = re.compile(r'\S+@\S+')
    text = EMAIL_RE.sub('', text)
    
    # remove attributions, eg.
    # RYAN MURRAY Bellevue Reporter Staff Writer
    # but leave 
    # - MISD Principals
    # - J. Jacob EdelMercer Island Reporter [???]
    words = text.split()
    first_word = words[0]
    if first_word.isupper() \
    and not first_word.endswith('.'):
        if len(words) > 1:
            second_word = words[1]
            if not (first_word.startswith('MI') \
                     and len(first_word) < 5 \
                    and not second_word.isupper()):
                keepers = []
                i = 0
                while i < len(words) and words[i].isupper():
                    keepers.append(words[i].capitalize())
                    i += 1
                text = ' '.join(keepers)
        
    # J. Jacob EdelMercer Island Reporter [???]
    text = text.replace('Mercer Island Reporter', '')
    
    # Leona Vaughn, WNPA News Service
    if ',' in text:
        text, _ = text.split(',')
    
    # Vicki Rackner, M.D.
    # Vicki Rackner, MD
    text = text.replace(' MD', ' M.D.')
    
    text = text.strip()
    
    if text != author and not author.lower().strip().startswith('by'):
        print(f'{author}\t===>\t{text}')
    
    if len(text) > max_len:
        print('Removing article:')
        print(author)
        return False
    
    return text

def get_links(base_url):
    page = 1
    while True:
        print(page)
        url = f'{base_url}/{page}'
        res = requests.get(url, headers=HEADERS)
        if res.status_code > 200:
            return
        soup = BeautifulSoup(res.text)
        entries = soup.find_all('header', {'class': 'entry-header'})
        for entry in entries:
            date = title = link = author = ''
            
            art = entry.find('span', {'class': 'entry-title'})
            title = art.text.strip()
            link = art.find('a')['href']
            
            byline = entry.find('div', {'class': 'latest-by'}).text.strip()
            if "•" in byline:
                author, date = byline.split("•")
                date = date.strip()
            else:
                date = byline
            
            if not author:
                author_parent = entry.find_next_sibling()
                author_block = author_parent.find('p').text.strip()
                if author_block.lower().startswith('by'):
                    author = author_block
                
            author_clean = clean_author(author)
            if author_clean != False:
                yield title, author_clean, link, date
        
        page += 1
        
def main():
    with open('article_data3.tsv', 'w') as outfi:
        for section in SECTIONS:
            print(section)
            base_url = f'{MI_REPORTER_BASE}/{section}/page'
            for (title, author, link, date) in get_links(base_url):
                outfi.write('\t'.join((author, section, date, title, link)))
                outfi.write('\n')
        
main()
# url = f'{MI_REPORTER_BASE}/pair-of-mercer-island-school-district-levy-renewals-are-passing-after-early-results/'    
# title, text = get_article_text(url)
# occurences = find_in_text(text, 'two')
# print(len(occurences))
```

```python
words_to_count = ['plethora', 'copious']

def get_article_text(url):
    html = requests.get(url, headers=HEADERS).text
    soup = BeautifulSoup(html)
    body = soup.body

    text = body.find('div', {'class': 'entry-content'}).text
    return text

def find_in_text(text, regex):
    text_lower = text.lower()
    return re.findall(regex, text)

with open('article_data3.tsv', 'r') as infi:
    with open('article_data_with_counts.tsv', 'w') as outfi:
        header = ['author', 'section', 'date', 'title', 'url']
        header.extend(words_to_count)
        outfi.write('\t'.join(header))
        outfi.write('\n')
        
        line = infi.readline()
        i = 0
        while line:
            if i % 10 == 0:
                print(i)
            items = line.rstrip().split('\t')
            text = get_article_text(items[-1])
            for w in words_to_count:
                num = len(find_in_text(text, w))
                items.append(str(num))
            outfi.write('\t'.join(items))
            outfi.write('\n')
            
            line = infi.readline()
            i += 1
```
